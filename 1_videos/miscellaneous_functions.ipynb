```python
def convert__hhmmss_to_scalar(time_str):
    unit = time_str.split(":")
    if len(unit) == 3:
        time_scalar = float(unit[0])*3600 + float(unit[1])*60 + float(unit[2])
    elif len(unit) == 2:
        time_scalar = float(unit[0])*60 + float(unit[1])
    return time_scalar
```

```python
def convert_time_to_frame(time_scalar):
    frames = time_scalar*50
    return frames
```

```python
def convert_hhmmss_to_frame(time_str):
    return convert_time_to_frame(convert__hhmmss_to_scalar(time_str))
    
```
